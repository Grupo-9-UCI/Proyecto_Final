## EDA - ETL

```python
# Lo primero es Importar las librerias que utilizaremos en el proceso de Extraccion, Transformacion y carga de datos (ETL).
import pandas as pd
import numpy as np 
```

```python
#Se realizara la carga de archivos valores separados por comas (CSV)
df_inputevents_cv= pd.read_csv('INPUTEVENTS_CV.csv')
df_inputevents_mv= pd.read_csv('INPUTEVENTS_MV.csv')
df_labevents= pd.read_csv('LABEVENTS.csv')
df_microbiologyevents = pd.read_csv('MICROBIOLOGYEVENTS.csv')
df_procedureevents_mv = pd.read_csv('PROCEDUREEVENTS_MV.csv')
df_procedures_icd = pd.read_csv('PROCEDURES_ICD.csv')
df_icustays = pd.read_csv('ICUSTAYS.csv')
df_outputevents = pd.read_csv('OUTPUTEVENTS.csv')
```

Se eliminan columnas inncesarias de los datasets

INPUTEVENTS_CV = se eliminan las columnas rate, rateuom, stopped, newbottle, originalamount, originalrate y originalsite. Los supuestos valores nulos son celdas en blanco se dejan como estan ya que no son un obstaculo para el analisis

```python
# calcular el porcentaje de valores nulos para cada columna
null_percentages = df_inputevents_cv.isnull().sum() / len(df_inputevents_cv)

# imprimir las columnas que tienen mÃ¡s del 50% de valores nulos
print(null_percentages[null_percentages > 0.5])
```

```python
df_inputevents_cv
```

```python
df_inputevents_cv.drop(columns=['rate', 'rateuom', 'stopped', 'newbottle', 'originalamount', 'originalrate', 'originalsite', 'originalrateuom'], inplace=True)
```

```python
#se crea nuevo df para con los cambios para nuestro trabajo sin afectar al original
df_inputevents_cv.to_csv("df_inputevents_cv.csv", index=False)
```
